# 과제2.
### 오픈 api에 포함된 "설명"(SMPL_DESC)을 분석하여, 맛집 설명에서 가장 유의미하게 많이 등장하는 5개 단어를 제시하라

## 2-1. kkma.pos 방법

```python
# dg_mat[0]['data'][180]['SMPL_DESC']
# dg_mat[0] 은 중구, [180]은 181번째 가게를 뜻한다.
```

```python
# 일단 모든 구들의 SMPL_DESC 를 하나로 합쳐주자.
dg_smpl = []
for i in range(0, 8):
    for j in range(0, int(dg_mat[i]['total'])) :
        dg_smpl.append(dg_mat[i]['data'][j]['SMPL_DESC']) # 리스트 형태로 합쳐준다.
```

```python
# print(df_smpl)
```

```python
# 위 방식만 하면 SMPL_DESC들이 각각의 인덱스로 들어갈 뿐 아직 전체가 하나의 문장이 되지는 않았기 때문에
# join : 리스트인 dg_mnu를 하나의 str타입으로 합쳐주고 그 사이엔 ', '를 넣는다.
dg_smpl = ', '.join(dg_smpl) 
```

```python
# kkma.pos를 이용해서 토큰화 한 것을 변수에 담아준다.
# 변수에 담지 않고 for, if문을 돌릴 경우 kkma.pos 함수 호출을 매번 하게 되어서 어마어마한 시간이 걸린다.
dg_smpl_token = kkma.pos(dg_smpl)
```

```python
# kkma.pos를 사용해 명사와 어근만 추출하자

dg_smpl_token_N = []
for i in range(0, len(dg_smpl_token)) :
    if dg_smpl_token[i][1] == 'NN' :
        dg_smpl_token_N.append(dg_smpl_token[i][0])
    elif dg_smpl_token[i][1] == 'NNB' :
        dg_smpl_token_N.append(dg_smpl_token[i][0])
    elif dg_smpl_token[i][1] == 'NNG' :
        dg_smpl_token_N.append(dg_smpl_token[i][0])
    elif dg_smpl_token[i][1] == 'NNM' :
        dg_smpl_token_N.append(dg_smpl_token[i][0])
    elif dg_smpl_token[i][1] == 'NNP' :
        dg_smpl_token_N.append(dg_smpl_token[i][0])
    elif dg_smpl_token[i][1] == 'NP' :
        dg_smpl_token_N.append(dg_smpl_token[i][0])
    elif dg_smpl_token[i][1] == 'XR' :
        dg_smpl_token_N.append(dg_smpl_token[i][0])
```

```python
# print(dg_smpl_token_N)
```

```python
# 토큰별(단어별) 카운팅!
from collections import Counter

dg_smpl_token_cnt = Counter(dg_smpl_token)
dg_smpl_token_cnt.most_common(10)
```

### hannanum.nouns를 사용해서 도출한 유의미한 단어 5가지는 아래와 같다.
## 1. 다양한
## 2. 전통
## 3. 한우
## 4. 국내산
## 5. 신선한

kkma.pos와 hannanum.nouns 두 방법 중 hannanum.nouns의 결과를 채택하기로 했다.
그 이유는 kkma.pos의 '찜', '갈비'는 정확히 어떤 음식을 가리키는지(갈비찜, 아구찜, 돼지갈비, 소갈비 등) 알 수 없기 때문이다.

위 단어들을 유의미하다고 판단한 이유는 아래와 같다.

1. **다양한** : 메뉴의 다양성을 홍보하기 위해 사용된 것으로 보인다. 현대사회에 들어서면서 다양성 존중에 대한 인식은 사회 전반에 자리잡았다. 한국 역시 과거 집단주의 사회에서 점차 개인주의 사회로 변화하고 있으며 이에 따라 '개인화'와 '개인 맞춤형'은 시대의 트렌드로 떠올랐다. 이는 사회가 개인의 취사선택을 보다 중요하게 생각하고 있고 더 많은 이들의 needs를 충족시키려 한다는 것을 시사한다. 따라서 식당 설명란에 있는 '다양한'이라는 키워드는 식당을 홍보하기에 충분히 유의미한 단어라고 볼 수 있다.


2. **전통** : 식당의 전통과 역사를 강조하는 단어다. 전통있는 식당은 그만큼 오래 가게를 운영하며 노하우가 축적되고 단골고객도 많을 것이기 때문에 중요한 단어라고 생각된다. 다만, '전통식당'은 요즘 트렌드를 따라가지 못한다는 느낌을 받을 수도 있다. 그러나 실제 MZ 세대들은 이른바 '할매입맛'을 존중하고 더 나아가 그러한 세대를 '할매니얼' 세대라고 부를 만큼 옛스러운 것을 또다른 개성과 취향이라고 생각하는 분위기이기 때문에 전통성을 강조한 식당도 MZ세대의 주목을 충분히 받고 있는 것으로 보인다.


3. **한우** : 1990년대에 시작된 신토불이 운동과 더불어 2003년과 2008년의 미국 광우병 파동으로 인해 한국인의 한우에 대한 관심은 과거와 비교할 수 없이 커졌다. 또한 본격적으로 한우를 브랜딩화 하고 국가와 지자체가 나서서 체계적으로 관리하면서 한우는 '명품'의 이미지를 가질 수 있게 되었다. 때문에 맛집 리스트 중 한우를 판매하는 식당이 많은 것은 어쩌면 당연한 일일 것이다.


4. **국내산** : 한우와 마찬가지로 신토불이 운동을 통해 국내산에 대한 한국인들의 애정은 더욱 커졌다. 특히 현지에서 나는 식자재일 수록 유통과정이 짧기 때문에 더 신선할 수 있다는 생각에 많은 이들이 국내산을 찾고있다. 또한 원산지표시법이 강화되고 정부의 관리 아래에 국내산 농산물에 대한 품질인증과 철저한 이력추적 덕분에 국내산 식품은 믿고 먹을 수 있다는 인식이 강해졌다. 따라서 대구시 맛집들 중에서도 국내산 재료를 사용하는 맛집이 많은 것으로 보인다.


5. **신선한** : 식재료의 신선도가 맛집 선정에 있어서 중요한 이유는 최근 식품 유통기업들의 서비스를 살펴보면 알 수 있다. 먼저, 마켓컬리는 풀콜드체인 시스템을 통해 식재료를 신선한 상태로 하룻밤만에 배송하는 서비스를 제공하고 있으며, 이를 통해 1조 유니콘 기업으로 성장할 수 있었다. 이에 질세라 쿠팡에서도 로켓프레시 서비스를 새롭게 도입해 신선한 식재료를 고객들에게 배송하고 있다. 이러한 사례들을 통해 식재료의 신선도는 맛집 선정에 있어서 충분히 유의미한 단어라고 볼 수 있다.
